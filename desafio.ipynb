### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
precio_ingresos_tienda = tienda["Precio"].sum()
precio_ingresos_tienda2 = tienda2["Precio"].sum()
precio_ingresos_tienda3 = tienda3["Precio"].sum()
precio_ingresos_tienda4 = tienda4["Precio"].sum()

print('El total de ingresos de la Tienda 1 es:',precio_ingresos_tienda)
print('El total de ingresos de la Tienda 2 es:',precio_ingresos_tienda2)
print('El total de ingresos de la Tienda 3 es:',precio_ingresos_tienda3)
print('El total de ingresos de la Tienda 4 es:',precio_ingresos_tienda4)
```

# 2. Ventas por categoría

```python
# Agrupar por categoría y contar las ventas en cada tienda
ventas_categoria_tienda1 = tienda.groupby('Categoría del Producto').size()
ventas_categoria_tienda2 = tienda2.groupby('Categoría del Producto').size()
ventas_categoria_tienda3 = tienda3.groupby('Categoría del Producto').size()
ventas_categoria_tienda4 = tienda4.groupby('Categoría del Producto').size()

# Imprimir los resultados
print("Ventas por categoría en Tienda 1:")
print(ventas_categoria_tienda1.to_string())
print("\nVentas por categoría en Tienda 2:")
print(ventas_categoria_tienda2.to_string())
print("\nVentas por categoría en Tienda 3:")
print(ventas_categoria_tienda3.to_string())
print("\nVentas por categoría en Tienda 4:")
print(ventas_categoria_tienda4.to_string())

def top_categorias(categorias, nombre_tienda):
    categorias_populares = categorias.sort_values(ascending=False).head(3)
    print(f"\nTop 3 categorías más populares ({nombre_tienda}):")
    for categoria, ventas in categorias_populares.items():
        print(f"- {categoria}: {ventas} ventas")
    print("\n")  # Añadir un salto de línea para mejor formato
    return categorias_populares

# Llamar a la función para cada tienda
top_categorias(ventas_categoria_tienda1, "Tienda 1")
top_categorias(ventas_categoria_tienda2, "Tienda 2")
top_categorias(ventas_categoria_tienda3, "Tienda 3")
top_categorias(ventas_categoria_tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
# prompt: En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

# Calcular la calificación promedio para cada tienda
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

# Imprimir los resultados
print(f"Calificación promedio de la Tienda 1: {calificacion_promedio_tienda1}")
print(f"Calificación promedio de la Tienda 2: {calificacion_promedio_tienda2}")
print(f"Calificación promedio de la Tienda 3: {calificacion_promedio_tienda3}")
print(f"Calificación promedio de la Tienda 4: {calificacion_promedio_tienda4}")
```

# 4. Productos más y menos vendidos

```python

def productos_mas_menos_vendidos(productos, nombre_tienda):

    productos_vendidos = productos.groupby('Producto').size()
    sorted_sales = productos_vendidos.sort_values(ascending=False)

    masvendido = sorted_sales.head(5)
    bottom_sellers = sorted_sales.tail(5)

    print(f"\nTop 5 productos más vendidos en {nombre_tienda}:")
    print(masvendido.to_string())

    print(f"\nTop 5 productos menos vendidos en {nombre_tienda}:")
    print(bottom_sellers.to_string())

productos_mas_menos_vendidos(tienda, "Tienda 1")
productos_mas_menos_vendidos(tienda2, "Tienda 2")
productos_mas_menos_vendidos(tienda3, "Tienda 3")
productos_mas_menos_vendidos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python

envio_promedio_tienda1 = tienda['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

# Imprimir los resultados
print(f"Costo de envío promedio de la Tienda 1: {envio_promedio_tienda1}")
print(f"Costo de envío promedio de la Tienda 2: {envio_promedio_tienda2}")
print(f"Costo de envío promedio de la Tienda 3: {envio_promedio_tienda3}")
print(f"Costo de envío promedio de la Tienda 4: {envio_promedio_tienda4}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualización 1: Ingresos totales por tienda (gráfico de pastel)
ingresos_por_tienda = {
    'Tienda 1': precio_ingresos_tienda,
    'Tienda 2': precio_ingresos_tienda2,
    'Tienda 3': precio_ingresos_tienda3,
    'Tienda 4': precio_ingresos_tienda4
}

tiendas = list(ingresos_por_tienda.keys())
ingresos = list(ingresos_por_tienda.values())

plt.figure(figsize=(8, 6))
plt.pie(ingresos, labels=tiendas, autopct='%1.1f%%', colors=['#66b3ff', '#ff9999', '#99ff99', '#ffcc99'])
plt.title("Distribución de ingresos por tienda")
plt.axis('equal')
plt.show()


# Visualización 2: Ventas por categoría (gráfico de líneas)
def visualizar_ventas_categoria(ventas, nombre_tienda):
    plt.figure(figsize=(10, 6))
    ventas.plot(kind='line', marker='o', color='darkorange')
    plt.title(f'Ventas por categoría en {nombre_tienda}')
    plt.xlabel('Categoría del Producto')
    plt.ylabel('Número de ventas')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

visualizar_ventas_categoria(ventas_categoria_tienda1, "Tienda 1")
visualizar_ventas_categoria(ventas_categoria_tienda2, "Tienda 2")
visualizar_ventas_categoria(ventas_categoria_tienda3, "Tienda 3")
visualizar_ventas_categoria(ventas_categoria_tienda4, "Tienda 4")


# Visualización 3: Calificación promedio por tienda (gráfico de violín)
calificaciones_promedio = {
    'Tienda 1': calificacion_promedio_tienda1,
    'Tienda 2': calificacion_promedio_tienda2,
    'Tienda 3': calificacion_promedio_tienda3,
    'Tienda 4': calificacion_promedio_tienda4
}

tiendas = list(calificaciones_promedio.keys())
calificaciones = list(calificaciones_promedio.values())

plt.figure(figsize=(8, 6))
sns.violinplot(data=calificaciones, palette='pastel')
plt.xticks([0, 1, 2, 3], tiendas)
plt.xlabel("Tienda")
plt.ylabel("Calificación promedio")
plt.title("Distribución de calificaciones promedio por tienda")
plt.ylim(0, 5)
plt.show()


# Visualización 4: Costo de envío promedio por tienda (gráfico de puntos)
envios_promedio = {
    'Tienda 1': envio_promedio_tienda1,
    'Tienda 2': envio_promedio_tienda2,
    'Tienda 3': envio_promedio_tienda3,
    'Tienda 4': envio_promedio_tienda4
}

tiendas = list(envios_promedio.keys())
envios = list(envios_promedio.values())

plt.figure(figsize=(8, 6))
plt.scatter(tiendas, envios, color='mediumslateblue', s=200)
plt.xlabel("Tienda")
plt.ylabel("Costo de envío promedio")
plt.title("Costo de envío promedio por tienda")
plt.grid(True)
plt.show()


```
